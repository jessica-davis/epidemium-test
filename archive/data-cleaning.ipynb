```python
import pandas as pd
df_forecast = pd.read_csv('https://raw.githubusercontent.com/cdcepi/FluSight-forecast-hub/refs/heads/main/model-output/FluSight-ensemble/2025-12-13-FluSight-ensemble.csv')

df_forecast = df_forecast[(df_forecast['target'] == 'wk inc flu hosp') & (df_forecast['location'] == 'US')]

df_target = pd.read_csv("https://raw.githubusercontent.com/cdcepi/FluSight-forecast-hub/refs/heads/main/target-data/time-series.csv")

df_target = df_target[(df_target['target'] == 'wk inc flu hosp') & (df_target['location'] == 'US') & (df_target['target_end_date'] >= '2025-11-01') & (df_target['as_of']== '2026-02-04')] 
df_target = df_target[df_target['target_end_date'] < df_forecast.target_end_date.min()]

df_forecast.to_csv('forecast.csv', index=False)
df_target.to_csv('target.csv', index=False) 
```
