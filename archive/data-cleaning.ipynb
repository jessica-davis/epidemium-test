```python
import pandas as pd
df_forecast = pd.read_csv('https://raw.githubusercontent.com/cdcepi/FluSight-forecast-hub/refs/heads/main/model-output/FluSight-ensemble/2025-12-13-FluSight-ensemble.csv')

df_forecast = df_forecast[(df_forecast['target'] == 'wk inc flu hosp') & (df_forecast['location'] == 'US')]

df_target = pd.read_csv("https://raw.githubusercontent.com/cdcepi/FluSight-forecast-hub/refs/heads/main/target-data/time-series.csv")

df_target = df_target[(df_target['target'] == 'wk inc flu hosp') & (df_target['location'] == 'US') & (df_target['target_end_date'] >= '2025-11-01') & (df_target['as_of']== '2026-02-04')] 
df_target = df_target[df_target['target_end_date'] < df_forecast.target_end_date.min()]

df_forecast.to_csv('forecast.csv', index=False)
df_target.to_csv('target.csv', index=False) 
```

```python
import os
import pandas as pd
```

```python
files = os.listdir("../../FluSight-forecast-hub/model-output/FluSight-ensemble/")
files = sorted([f for f in files if pd.to_datetime(f.split("-Flu")[0]) >= pd.to_datetime("2024-10-01") and pd.to_datetime(f.split("-Flu")[0]) <= pd.to_datetime("2025-06-30")])

df_all_h0 = []
for file in files:
    df_forecast = pd.read_csv(f"../../FluSight-forecast-hub/model-output/FluSight-ensemble/{file}")
    df_forecast = df_forecast[(df_forecast['target'] == 'wk inc flu hosp') & (df_forecast['location'] == 'US')]

    df = df_forecast[df_forecast.horizon==0]
    df_all_h0.append(df.copy())

df_all_h0 = pd.concat(df_all_h0, ignore_index=True)
df_all_h0.to_csv('forecast_h0.csv', index=False)
```

```python
df_target = pd.read_csv("https://raw.githubusercontent.com/cdcepi/FluSight-forecast-hub/refs/heads/main/target-data/time-series.csv")
df_target = df_target[(df_target['target'] == 'wk inc flu hosp') & (df_target['location'] == 'US') & (df_target['target_end_date'] >= '2024-11-01') & (df_target['target_end_date'] <= '2025-05-31')]
df_target = df_target[df_target['as_of']== '2026-02-04']
df_target.to_csv('target_h0.csv', index=False)
```
